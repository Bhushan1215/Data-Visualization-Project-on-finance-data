```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("/kaggle/input/finance/finance.csv")
print(df.shape)
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df.dtypes)
```

```python
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
df['Debt_Revenue_Ratio'] = df['Totals. Debt at end of fiscal year'] / df['Totals.Revenue']
```

```python
# Visualization 1: Revenue vs Expenditure Trend (Nationwide)
plt.figure(figsize=(12,6))
nationwide = df.groupby(df['Year'].dt.year)[['Totals.Revenue', 'Totals.Expenditure']].sum()
plt.plot(nationwide.index, nationwide['Totals.Revenue'], label='Revenue')
plt.plot(nationwide.index, nationwide['Totals.Expenditure'], label='Expenditure')
plt.title('Nationwide Revenue vs Expenditure Trend')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.show()
```

**Insights from this Visualization**

Long-Term Trends: Shows whether revenue and expenditure are increasing or decreasing over the years.

Economic Events:

A drop in revenue around 2008 would show the impact of the financial crisis.

Any sudden spikes in expenditure may reflect stimulus spending or policy changes.

Budget Health:

If expenditure exceeds revenue, it indicates a deficit.

If revenue is greater than expenditure, it suggests a surplus.



```python
# Visualization 2: State Revenue Comparison (2019)
plt.figure(figsize=(14,8))
df_2019 = df[df['Year'].dt.year == 2019].sort_values('Totals.Revenue', ascending=False)
sns.barplot(x='Totals.Revenue', y='State', data=df_2019.head(15), palette='viridis')
plt.title('Top 15 States by Revenue (2019)')
plt.xlabel('Revenue (Billions)')
plt.ylabel('State')
plt.show()
```

**Insights from this Visualization**

 1. California Leads Significantly

    California is at the top by a wide margin.
 
    Its high revenue is driven by:

 2. A large population

    High personal income tax and corporate tax collections

    A strong tech-based economy

```python
pivot = df.pivot_table(index="State", columns="Year", values="Totals.Revenue")
sns.heatmap(pivot, cmap='YlGnBu')
plt.title("Revenue Heatmap by State & Year")
plt.show()
```

**Insights from this Visualization**

1.  **Growth Over Time**
For most states, colors darken left to right, showing revenue increases from early to recent years.

This suggests that state revenue has grown steadily over time.

2.  **Top Revenue States Pop Out**
California, New York, Texas etc. have darker shades consistently, indicating high annual revenue.

These states likely have stronger economies, larger populations, or more aggressive tax structures.

3.  **Recession Periods Are Visible**
Around 2008–2009, some states might show a lighter color (temporary revenue dip) due to the Great Recession.

This reflects how economic downturns affect government income.

4.  **Low-Revenue States**
States like Wyoming, Vermont, Mississippi tend to show lighter shades consistently, indicating lower revenues.

Often these are less populous or rural states with smaller tax bases.



```python
# Visualization 4: Expenditure Composition (California 2019)
ca_2019 = df[(df['State'] == 'CALIFORNIA') & (df['Year'].dt.year == 2019)]
exp_cols = ['Details.Education.Education Total', 'Details.Health.Health Total Expenditure',
            'Details.Transportation.Highways.Highways Total Expenditure', 
            'Details.Welfare.Welfare Institution Total Expenditure']
exp_values = ca_2019[exp_cols].squeeze()
plt.figure(figsize=(10,8))
plt.pie(exp_values, labels=exp_cols, autopct='%1.1f%%', startangle=90)
plt.title('California Expenditure Composition (2019)')
plt.show()
```

**Insights from this Visualization:-**

  1 **Expenditure Priorities Are Clearly Visible**
The chart shows which sectors consume the most government funding in California.

For example, if education or health takes up a large slice (e.g., 40–50%), it reflects state policy priorities.

  2. **Education and Health Typically Dominate**
In most states — especially California — education and healthcare are likely to take the largest shares.

This reflects large investments in public schools, universities, Medicaid, and public health programs.

  3. **Welfare and Highways Receive Smaller Shares**
Transportation (Highways) and Welfare Institutions often receive a smaller proportion.

However, these may still represent billions in actual dollars, even if their % share looks small.

  4. **Comparative Use**
This pie chart can be compared with other states to see how California differs in budget allocation.

For instance, Texas or Florida may have different weightings, especially in health or welfare.

```python
# Visualization 5: Capital Outlay Trends
plt.figure(figsize=(12,6))
top_states = df.groupby('State')['Totals.Revenue'].sum().nlargest(5).index
for state in top_states:
    state_data = df[df['State'] == state]
    plt.plot(state_data['Year'].dt.year, state_data['Totals.Capital outlay'], label=state)
plt.title('Capital Outlay Trends in Top 5 States by Revenue')
plt.xlabel('Year')
plt.ylabel('Capital Outlay (Millions)')
plt.legend()
plt.grid(True)
plt.show()
```

**Insights From this Visualization:-**

    Capital outlay represents long-term investments such as:

    Roads, bridges

    Schools, hospitals

   Public buildings and large-scale infrastructure

**A rise in capital outlay indicates economic development initiatives, infrastructure upgrades, or stimulus programs.**

 2. **Economic Cycles Influence Investment**
You may observe dips during economic crises (e.g., 2008 recession), and spikes post-recovery.

States may reduce capital spending during downturns and increase during growth periods or stimulus years.

 3. **California Likely Leads in Capital Outlay**
Due to its large economy and population, California often shows consistently high and rising infrastructure investment.

 4.**State-by-State Policy Differences**
Some states might show volatile spending, indicating inconsistent infrastructure strategies.

Others show steady upward trends, suggesting long-term planning and growth.

 **5. Comparative Efficiency**
Comparing trends helps assess:

Which states prioritize infrastructure more

```python
# Visualization 6: Revenue Sources Comparison
rev_cols = ['Totals.Intergovernmental', 'Totals.Tax', 'Totals.License tax', 
            'Totals.Selective sales tax']
rev_2019 = df[df['Year'].dt.year == 2019][rev_cols].sum()
plt.figure(figsize=(10,6))
sns.barplot(x=rev_2019.values, y=rev_2019.index, palette='mako')
plt.title('National Revenue Sources (2019)')
plt.xlabel('Amount (Billions)')
plt.ylabel('Revenue Source')
plt.show()
```

**Relative Contribution of Each Revenue Source:**

The length of each bar shows how much that source contributed in 2019.

You can immediately tell which sources generated more or less revenue.

**Tax Dominance**

Likely, Totals.Tax has the largest bar, indicating that general taxation is the primary source of national revenue.

This includes income taxes, property taxes, etc.

```python
# Visualization 7: Debt vs Expenditure Correlation
plt.figure(figsize=(10,8))
sns.scatterplot(x='Totals.Expenditure', y='Totals. Debt at end of fiscal year', 
                size='Totals.Revenue', hue='Totals.Revenue',
                data=df[df['Year'].dt.year > 2010], sizes=(20, 200), alpha=0.7)
plt.xscale('log')
plt.yscale('log')
plt.title('Debt vs Expenditure (2011-2019)')
plt.xlabel('Expenditure (Log Scale)')
plt.ylabel('Debt (Log Scale)')
plt.show()
```

**. Strong Positive Correlation Between Expenditure and Debt**
The overall pattern likely shows an upward slope, even on a log-log scale—meaning:

States that spend more also tend to carry more debt.

This is expected: higher government spending often coincides with greater borrowing needs, particularly for infrastructure, education, and healthcare.

```python
# Visualization 8: Education Expenditure Growth
plt.figure(figsize=(10,5))
education = df.groupby(df['Year'].dt.year)['Details.Education.Education Total'].sum()
education.plot(kind='line', marker='o')
plt.title('National Education Expenditure Trend (1992-2019)')
plt.xlabel('Year')
plt.ylabel('Education Expenditure (Billions)')
plt.grid(True)
plt.show()
```

1. **Spending Growth Is Not Uniform**
The line may show periods of slower or faster growth:

Plateaus or dips could correspond to economic downturns (e.g., early 2000s, post-2008 recession).

Steeper increases might align with economic booms or education reform efforts.

**2.   Supports Need for Continued Investment**
The visualization supports the narrative that education remains a growing national priority, though the rate and consistency of growth may vary by political and economic climate.

```python
# Visualization 9: Revenue Per Capita Comparison (2019)
# Adding mock population data for illustration
pop_data = {
    'CALIFORNIA': 39538223, 'TEXAS': 29145505, 'FLORIDA': 21538187,
    'NEW YORK': 19453561, 'PENNSYLVANIA': 12801989
}
df_2019['PerCapitaRevenue'] = df_2019.apply(
    lambda row: row['Totals.Revenue'] / pop_data.get(row['State'], 1), axis=1)
top_5_pop = df_2019[df_2019['State'].isin(pop_data.keys())]
plt.figure(figsize=(10,6))
sns.barplot(x='PerCapitaRevenue', y='State', data=top_5_pop.sort_values('PerCapitaRevenue', ascending=False))
plt.title('Per Capita Revenue Comparison (2019)')
plt.xlabel('Revenue Per Capita')
plt.ylabel('State')
plt.show()
```

**Per Capita Revenue is not equal Total Revenue:**

A state with a large total revenue (like California) might not have the highest per capita revenue if its population is also very large.

This normalization helps assess fiscal capacity per resident, not just raw financial size.

**State-by-State Comparison**c

You might find New York or California at the top, as they tend to have high per capita taxes and services.

Florida and Texas may rank lower due to no personal income tax and potentially lower tax burden
